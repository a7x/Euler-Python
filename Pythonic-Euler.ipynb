Problem 1
=========

If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

#### Previous solution 

```python
#Sum of all multiples of 3 or 5 below 1000
def ifDivisible(n):
        #return true if divisible by 3 or 5
        if(n%3==0 or n%5==0):
                return True
        else:
                return False
sum = 0
for i in range(3,1000):
        if(ifDivisible(i)):
                sum=sum+i
print sum
```

Quick, what are the problems with the code above ? 

0. Well, sum is a keyword ! dont use keywords, ever 

1. don't use range ! it builds up a list in memory. use xrange ! 

2. if its not true, its false. no need of the else condition here 

3. is it really necessary to sum over the way its been done ?

```python
def ifDivisible(n):
    if n%3 == 0 or n%5 == 0:
        return True
    return False

print sum([i for i in xrange(3, 1000) if ifDivisible(i)])
```

I did not have a previous solution for this one. so here goes one now. 

```python
#find the 10001st prime number 
import math
def ifPrime(n):
    ''' returns true/false w.r.t primality of a number. 
        From wikipedia, we find that all primes are of the form 6k +/- 1
        So, instead of doing it till sq(n), lets check, 2, 3 
        and then every number of the form 6k +/- 1
    '''
    
    sqrt = math.sqrt(n)
    if n == 2 or n == 3 or n == 5:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    if str(n).endswith('0') or str(n).endswith('5'):
        return False

    
    k = 1
    first, second = (6 * k) - 1, (6 * k) + 1
    while first <= sqrt:
        if n%first == 0 or n%second == 0:
            return False
        k += 1
        first, second = (6 * k) - 1, (6 * k) + 1
    return True

num_primes = 6
start = 17
while True:
    if num_primes == 10001:
        print start - 2
        break
    
    if ifPrime(start):
        num_primes += 1
    start += 2
```
